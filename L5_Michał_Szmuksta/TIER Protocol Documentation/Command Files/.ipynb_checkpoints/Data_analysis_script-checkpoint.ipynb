```python
import pandas as pd
from pandasql import sqldf
import matplotlib.pyplot as plt

Drinks = pd.read_csv("../Analysis Data/drinks_final.csv")
```

**Statystyki**
Średnie spożycie alkoholi na świecie:

```python
Df = sqldf("SELECT continent, AVG(beer_servings), AVG(spirit_servings), AVG(wine_servings), AVG(total_litres_of_pure_alcohol) FROM Drinks "
           "GROUP BY continent")
Df
```

```python
fig, ax = plt.subplots(figsize=(6, 6))
Df.plot.pie(ax=ax, y='AVG(total_litres_of_pure_alcohol)', labels=['Afryka','Azja','Europa','Ameryka Północna','Oceania','Ameryka Południowa'])
ax.set_title('Średnie roczne spożycie alkoholu przez osobę na danym kontynencie')
ax.get_legend().remove()
ax.set_ylabel('')
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(1,3, figsize=(16, 6))
Df.plot.pie(ax=ax[0], y='AVG(beer_servings)', labels=['Afryka','Azja','Europa','Ameryka Północna','Oceania','Ameryka Południowa'])
ax[0].set_title('Średnie roczne spożycie piw przez osobę')
ax[0].get_legend().remove()
ax[0].set_ylabel('')
plt.tight_layout()

Df.plot.pie(ax=ax[1], y='AVG(spirit_servings)', labels=['Afryka','Azja','Europa','Ameryka Północna','Oceania','Ameryka Południowa'])
ax[1].set_title('Średnie roczne spożycie alkoholi spirytusowych przez osobę')
ax[1].get_legend().remove()
ax[1].set_ylabel('')
plt.tight_layout()

Df.plot.pie(ax=ax[2], y='AVG(wine_servings)', labels=['Afryka','Azja','Europa','Ameryka Północna','Oceania','Ameryka Południowa'])
ax[2].set_title('Średnie roczne spożycie win przez osobę')
ax[2].get_legend().remove()
ax[2].set_ylabel('')
plt.tight_layout()

plt.show()
```

10 krajów z największym spożyciem alkoholu na osobę:

```python
Df = sqldf("SELECT country, total_litres_of_pure_alcohol FROM Drinks ORDER BY total_litres_of_pure_alcohol DESC LIMIT 10")
Df
```

Liczba krajów z zerowym spożyciem alkoholu:

```python
Df = sqldf("SELECT COUNT(country) FROM Drinks WHERE total_litres_of_pure_alcohol = 0")
Df
```
